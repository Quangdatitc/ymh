```python
def ProcessingSTFT(xlim=[0, 0.5], ylim=[0, 22.05], clim=[0, 500], AxisSetting = False, Resizing = True): # 秒  # 周波数[kHz] # 強度)
    import numpy as np
    import os
    import pandas as pd

    InputDataPath = r".\data"
    ResultPath = r'.\result'

    NFFT = 256 # STFTのデータ数/1フレーム = 1画像中のデータ数

    ###################STFT画像の表示方法の設定################################
    # Documents/Python Scripts/sound/STFT.py
    # xlim=[0, 0.5]
    # ylim=[0, 5]
    # clim=[0, 20]

    ResultPath = ResultPath + '-' + str(xlim[-1]) + '-' + str(ylim[-1]) + '-' + str(clim[-1])
    if os.path.exists(ResultPath): pass
    else: os.makedirs(ResultPath)

    ############トリミング後のサイズ##############################
    size=128
    save_kaisuu=np.arange(0,9.5,10) # 画像化の回数
    # save_kaisuu=np.arange(0,9.5,0.1)

    ############データ読込##############################

    import os
    import glob
    #Documents/Python Scripts/sound/list/220228_OP10 2回目.wav
    #Documents/Python Scripts/sound/data/PROJ000/AUDIO/MONO-000.WAV
    InputDatas = ['data/220601_0085 NG1.wav', 'data/220601_0086 NG2.wav','data/220601_0087 OK1.wav','data/220601_0088 OK2.wav','data/220601_0089 OK BRKT.wav','data/220601_0090 NG BRKT.wav']
    #print('InputDatas',InputDatas)
    Waves = [os.path.abspath(s) for s in InputDatas] # 音声データ絶対パス
    InputIndex = [s[:7] for s in InputDatas] # データのインデックス
    print(Waves)
    ############STFT##############################
    import wave
    import math
    import matplotlib.pyplot as plt
    from PIL import Image, ImageOps

    for k, path in enumerate(Waves):
        if k == 3 or k == 10 :
            pass
        WavesMaxC = 0

        wf=wave.open(path,"r")
        x=os.path.basename(path)
        buf=wf.readframes(wf.getnframes())
        sampling_rate=wf.getframerate() #wavファイルのサンプリングレート = サンプル数/秒 通常44.1kHz = 44100
        data = np.frombuffer(buf, dtype='int16')/32768.0
        ms=np.mean(data**2)
        rms=math.sqrt(ms)

        frame_length = data.shape[0] # wavファイルの全フレーム数
        time_song = float(frame_length) / sampling_rate  # 波形長さ(秒)
        time_unit = 1 / float(sampling_rate) # 1サンプルの時間(秒)
        OVERLAP = int(NFFT / 2.0)#フレームの重なりを指定
        start = (NFFT / 2) * time_unit # 開始位置 (秒) 128データは除外
        stop = time_song # 終了位置 (秒)
        step =  (NFFT - OVERLAP) * time_unit # 次の画像との時間差 (秒)
        time_ruler = np.arange(start, stop, step)


        window = np.hamming(NFFT)#窓関数の指定

        spec = np.zeros([len(time_ruler), int(1 + (NFFT / 2))]) #転置状態で定義初期化
        pos=0

        ###########################STFTを実行####################################
        for fft_index in range(len(time_ruler)):
            frame = data[pos:pos+NFFT]
            if len(frame) == NFFT:
                windowed = window * frame
                fft_result = np.fft.rfft(windowed)
                fft_data = np.abs(fft_result) ** 2
                for i in range(len(spec[fft_index])):
                    spec[fft_index][-i-1] = fft_data[i]
                pos += (NFFT - OVERLAP)

        for sv in save_kaisuu:
            print(f'{sv:.1f} / {save_kaisuu[-1]} \r', end = '')
        ###################作成したデータの保存先の指定#########################
            name=os.path.dirname(os.path.dirname(path))#wavファイルの保存されているフォルダのパスを取得
            name1=os.path.basename(name)#STFTするデータのシリアルNoを取得
            name2=name1.translate(str.maketrans({chr(0xFF01 + i): chr(0x21 + i) for i in range(94)}))#全角を半角に変換
            SAVE_DIR = ResultPath
            sv=round(sv,1)
            
            save_name = "{}\\{}_{}.png".format(name2,k,sv)
     
            # 2022年5月17日　保存名称変更
            print(f'savename={save_name} -> {x}')
            save_name = f'{name2}\\{x}.png'
            
            save_path = ResultPath + "\\"+ save_name
            
            #　フォルダ有無確認
            SaveFolder = os.path.dirname(save_path)
            if os.path.exists(SaveFolder) : pass
            else: os.mkdir(SaveFolder)

            #スペクトログラムの表示設定
            plt.figure(figsize=(11,8))
            plt.imshow(spec.T, extent=[0, time_song, 0, sampling_rate/2/1000], aspect="auto",cmap="jet",interpolation='none', origin = 'upper', ) # x=time, y =frequency でマップ化
            plt.xlim(xlim[0]+sv,xlim[1]+sv)
            plt.ylim(ylim)
            plt.clim(clim)
            if AxisSetting:
                plt.colorbar(aspect=40, pad=0.08, shrink=1.0, orientation='vertical', extend='neither', extendrect = True)
                plt.xlabel('Time[s]')
                plt.ylabel('Frequency[kHz]')
            else:
                plt.axis("off")
                plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
            plt.savefig(save_path)
            
            plt.clf()
            plt.cla()
            plt.close()

            ###STFT画像のトリミング        
            im = Image.open(save_path).convert('RGB')#読み込んだフォルダのjpgファイルのパスの配列の1番目のデータを開く
            plt.imshow(im)
            invert_im = ImageOps.invert(im)#開いたデータの画素値を逆転

            im = Image.open(save_path, "r")
            img=im.crop(invert_im.getbbox())
            im_resize = im.resize((size, size),Image.NEAREST)
            
            if Resizing: im_resize.save(save_path)

            maxim = np.max(spec)
            if WavesMaxC < maxim: WavesMaxC = maxim
        print(f'{k} / {len(Waves)} , C_max = {WavesMaxC}')

    print("###end###")
```

```python
#import STFT
x = [0, 20]
y = [0,6]
c = [0, 3000]
ProcessingSTFT(xlim = x, ylim = y, clim = c, AxisSetting=True, Resizing = False)
```
